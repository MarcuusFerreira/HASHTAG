## Formatação de strings

```python
email = str('marcusvcf04@gmail.com')

print(f'Meu e-mail não é {email:^30}, show?')
```

```python
custo = int(500)
faturamento = int(270)
lucro = int(faturamento - custo)

print(f'Faturamento foi {faturamento:+} e lucro foi {lucro:+}')
```

```python
custo = int(5000)
faturamento = int(2700)
lucro = int(faturamento - custo)

print(f'Faturamento foi {faturamento:+,} e lucro foi {lucro:+,}')
```

## Formatação de moeda

```python
custo = int(5000)
faturamento = int(27000)
lucro = int(faturamento - custo)

print(f'Faturamento foi R${faturamento:,.2f} e lucro foi R${lucro:,.2f}')

# formato brasileiro
lucro_texto = 'R${:_.2f}'.format(lucro)
print(lucro_texto.replace('.', ',').replace('_', '.'))
```

## Função round()

```python
imposto = 0.15758
preco = 100
valor_imposto = round(preco * imposto, 1)
print(f'Imposto sobre o preço é de R${valor_imposto}')
```
